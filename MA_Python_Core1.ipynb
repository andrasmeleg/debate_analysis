```python
def move_2_project_folder(filename):         #makes a given file to a project folder where I collect all html files
    current_working_directory = os.getcwd()
    shutil.move(current_working_directory+"/"+filename, current_working_directory+"/"+HTML_files+"/"+filename)

        
def writing_html_file_from_source(filename, source):
    file_ = open(filename, "w")
    file_.write(source)
    file_.close()
    print(filename)                          #So, what is our filename? 
    move_2_project_folder(filename)          #It is time to move the html file to a folder created for html files

def save_html(filename, url):
    """
    It saves an html file, to a given string filename, if you give the url, which you want to save. 
    """
    print(url)                                #what is the input url?
    response = requests.get(url,verify=False)              #request the webpage to be opened for reading
    print(response.status_code)               #hopefully 200 each --> successful download
    html_source = response.content.decode('utf-8')       #decoding the file html to a txt
    writing_html_file_from_source(filename, html_source) #puts the response to the project's html folder as a file
    return response
```

```python
class Debate_Team:
    """
    This is a Debate Team Class
    """
    
    def __init__(self, university, speakers):
        
        self.speakers = speakers
        self.university = university
        self.team_ELO = 1000
        
    def print_speakers(self):
        """
        Print the Debate_Team's speakers"""
        for speaker in self.speakers:
            print(speakers)
    def print_university(self):
        """
        Print the Debate_Team's university"""
        print(self.university)
    
    #https://metinmediamath.wordpress.com/ + next line 
    #2013/11/27/how-to-calculate-the-elo-rating-including-example/
    #https://dotesports.com/general/news/elo-ratings-explained-20565
    #https://en.wikipedia.org/wiki/Elo_rating_system
    def change_team_ELO(self, other, result):
        """Changes the elo of the team after a debate"""
        home_ELO = self.team_ELO
        away_ELO = other.team_ELO
        result = result
        Expected_score_home = 1 / (1 + 10**((away_ELO-home_ELO)/400))
        K_factor = 16
        score = 1*result
        new_home_ELO = home_ELO + K_factor*(score - Expected_score_home)
        self.ELO = new_home_ELO
        
        return self.ELO
```

```python
#The next 2 functions are for generating soup for data analysis, same input, same output, different solution

#This function returns the soup of the html file with selenium 
#The input "team" is from a previous element of the program.
def open_team_webdriver_html_tabbie(team, toprint):
    link= "https://tabbie.org/" + team     #generating the link
    # print(link)                            #check if we have good link
    driver.get(link)                       #getting html opened in driver
    html_source = driver.page_source       #preparing source input
    soup = BeautifulSoup(html_source, 'html.parser') 
    if toprint == True:                    #In case I want to print out the content 
        print("html open:" + team)
    return soup

#This function returns the soup of the html file from local files
#The input "team" is from a previous element of the program.
def open_team_local_html_tabbie(team, toprint):
    name = team.replace("/","-") + ".html"
    local_html_link = os.getcwd() + "/" + HTML_files + "/" + name
    soup = BeautifulSoup(open(local_html_link, encoding = "utf-8"), "html.parser")
    if toprint == True:
        print("html open:" + team)
    return soup

#This is to get a text from between the html brackets such as: <example/> --> example
def get_text_from_html_brackets(text_with_brackets, toprint):
    twb = text_with_brackets
    twb = str(twb)
    result = twb[4:-5]
    if toprint == True:
        print(result)
    return result

#Another one of this, for different bracket types, such as: <example> --> example
def get_text_from_html_brackets_alternative(text_with_brackets, toprint):
    twb = text_with_brackets
    twb = str(twb)
    result = twb[3:-4]
    if toprint == True:
        print(result)
    return result

#It gets an ID of the team, such as: Vienna-EUDC-2015/team/450 --> 450 
#The ID is used as identification key in the database. 
def get_ID(team, toprint):
    ID = team[-5:]
    if ID[0] == "m":
        ID = team[-3:]
    if ID[0] == "/":
        ID = team[-4:]
    if toprint == True:
        print("ID of this team: ",ID)
    return ID

#This function returns the teamname from the tabbie_html, if the soup input is given
def get_team_name_from_tabbie_team_html(soup, toprint):
    tn_in_brackets = soup.find("h1")
    teamname = get_text_from_html_brackets(tn_in_brackets, toprint)
    return teamname

#This function returns team data in list from the tabbie_html, if the soup input is given
#The elements of the list are: [Spekarname_A, Speakername_B, team_language_status, team_points, team_speakerpoints]
def getting_speaker_names_from_tabbie_team_html(soup, toprint):
    counter = 0
    for tables in soup.find_all("table"):
        for table in tables:
            line = soup.find_all("td")
            speaker_A = get_text_from_html_brackets(line[0],False)
            speaker_B = get_text_from_html_brackets(line[1],False)
            language_status = get_text_from_html_brackets(line[2],False)
            points = get_text_from_html_brackets(line[3],False)
            speaks = get_text_from_html_brackets(line[4],False) 
            counter += counter 
    result = [speaker_A , speaker_B, language_status, points, speaks]
    if toprint == True:
        print(result)
    return result

#This function returns a list, which contains 36 ID numbers of teams. Its structure goes as the following: 
#Last round of the debate tournament: OG_ID, OO_ID, CG_ID, CO_ID
#Second-last round of the debate tournament: OG_ID, OO_ID, CG_ID, CO_ID
#etc.
def getting_results_with_ID_numbers_from_team_html(soup, toprint):
    ID_numbers = []
    for url in soup.find_all('a'):
        link = url.get('href')
        ID_numbers.append(link)
    ID_numbers = ID_numbers[10:-10]
    counter = 0
    for ID_number in ID_numbers:
        ID_number = ID_number[-5:]
        if ID_number[0] == "m":
            ID_number = ID_number [-4:]
        if ID_number[0] == "/":
            ID_number = ID_number [-3:]
        ID_numbers[counter] = ID_number
        counter += 1 
    if toprint == True:
        for element in ID_numbers:
            print(element)
        print("elements:" + str(len(ID_numbers)))
    return ID_numbers

#This returns a list, with the chairs, rounds, and ranking results. It does not contain team ID-s, only results. 
def getting_rankings_chairs_from_tabbie(soup, toprint):
    rankings_and_chairs_and_rounds_tabbie=[]
    for rank in soup.find_all("b"):
        info = get_text_from_html_brackets_alternative(rank, False)
        rankings_and_chairs_and_rounds_tabbie.append(info)
    r_a_c_a_r_t = rankings_and_chairs_and_rounds_tabbie 
    rankings_and_chairs_and_rounds_tabbie = r_a_c_a_r_t[2:-1]
    if toprint == True:
        for elements in rankings_and_chairs_and_rounds_tabbie:
                print(elements)
    result = rankings_and_chairs_and_rounds_tabbie
    return result
```

```python
debate_teams = [] #Supposed structure: list [elements]
#element = {team_ID : teamname}
debaters = [] #Supposed structure: list [elements] 
#element = { team_ID : [debatername_A, debatername_B] } 
language_status = [] #element = {team_ID : langueage status}
#language status in debating can be: EPL / ESL (Englsih as Primary Language, English as Secondary Language)
point_results = [] #element = {team_ID : point results of team}
speaker_points = [] #element = {team_ID : speaker points of team}
all_results = [] #it has a complex data structure hard to explain... 
#see function: getting_results_with_ID_numbers_from_team_html
ID_results = [] #this has also a complicated structure
#see function: getting_rankings_chairs_from_tabbie
```
